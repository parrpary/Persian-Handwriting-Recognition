```python
import numpy as np
import cv2
import cv2, PIL
from cv2 import aruco
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
import glob
from operator import itemgetter
from PIL import Image
from PIL import ImageFilter
from matplotlib import pyplot as plt
```

```python
### --------------- FIRST PAGES --------------- ###
turn = 0
# print("turn_1",turn)

pathh = 'rawdataset/'
fnames = glob.glob(os.path.join(pathh, "*_1*.jpg"))
# print("fnames",fnames)

for filename in fnames:
    frame = cv2.imread(filename)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     print("raw image \n")
#     cv2.imshow('gray',gray)
    
    ## Detecting Markers ##
    
    aruco_dict = aruco.Dictionary_get(aruco.DICT_6X6_250)
    parameters =  aruco.DetectorParameters_create()
    corners, num, rejectedImgPoints = aruco.detectMarkers(gray,aruco_dict,parameters=parameters)
    frame_markers = aruco.drawDetectedMarkers(frame.copy(), corners, num)
    #res_corners = np.concatenate(corners, axis = 1)
    
    samp = np.zeros([4,2], dtype=np.float32)
#     print("detecting \n")
#     plt.figure()
#     plt.imshow(frame_markers)
    for i in range(len(num)):
        c = corners[i][0]
        eps = 0
        if num[i]==30:
            samp[0] = (c[0][0]+c[0][0]*eps,c[0][1]-c[0][1]*eps)
        if num[i]==31:
            samp[1] = (c[1][0]+c[1][0]*eps,c[1][1]+c[1][1]*eps)
        if num[i]==33:
            samp[2]= (c[2][0]-c[2][0]*eps,c[2][1]+c[2][1]*eps)
        if num[i]==32:
            samp[3] = (c[3][0]-c[3][0]*eps,c[3][1]-c[3][1]*eps)
        #plt.plot([c[:, 0].mean()], [c[:, 1].mean()],"o", label = "id={0}".format(num[i]))
#     plt.legend()
#     plt.show()

    ## Perspective ##
    n = 2562 #2480
    m = 1820 #1748
    points = np.array([(0,0),
                        (m,0),
                        (m,n),
                        (0,n)]).astype(np.float32)

    H = cv2.getPerspectiveTransform(samp,points)
    output_size = (m,n)
    perspective = cv2.warpPerspective(frame,H,output_size)
#     gray_pers = cv2.cvtColor(perspective, cv2.COLOR_BGR2GRAY)
#     print("perspective \n")
#     plt.imshow(perspective)
#     plt.show()
    
    ## Separating ##
    path_save = 'dataa/'
    for i in range(21):
        for j in range(10):
            if i == 0 or i == 1 or i == 19 or i == 20:
#                 cv2.circle(perspective, (250 + 130* j, 10+122*i), 20, (0, 0, 255),-1)
                x = 250
                w = 130
                h = 122
                y = 10
                ROI = perspective[y + i * h:y + h * (i + 1), x + j * w:x + (j + 1) * w]

                ROI = cv2.cvtColor(ROI, cv2.COLOR_BGR2GRAY)
                ROI = np.array(ROI,dtype=np.uint8)
                ROI = cv2.GaussianBlur(ROI, (3, 3), 0)
                _, ROI = cv2.threshold(ROI, 200, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
                cv2.imwrite(os.path.join(path_save, '{}_a/d{}.jpg'.format(i, (j + 10 * turn))), ROI)
                
        for j in range(14):
            if i != 0 and i != 1 and i != 19 and i != 20:
                # cv2.circle(perspective,(130*j,10+122*i),20,(0,255,255),-1)
                x = 0
                y = 10
                w = 130
                h = 122
                ROI = perspective[y + i * h:y + h * (i + 1), x + j * w:x + (j + 1) * w]


                ROI = cv2.cvtColor(ROI, cv2.COLOR_BGR2GRAY)
                ROI = np.array(ROI,dtype=np.uint8)
                ROI = cv2.GaussianBlur(ROI, (3, 3), 0)
                _, ROI = cv2.threshold(ROI, 200, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
                cv2.imwrite(os.path.join(path_save, '{}_a/d{}.jpg'.format(i, (j + 14 * turn))), ROI)

    turn = turn+1
#     print("turn_1",turn)
```
